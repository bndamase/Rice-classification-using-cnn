##**Blur Detection**

Image quality was assessed by measuring blurriness across the dataset, as low-quality images can reduce feature clarity and negatively affect CNN performance (Mahmud et al., 2021). The Laplacian Variance method was used to quantify image sharpness, where lower variance scores indicate a higher level of blur (Mahmud et al., 2021). Five sample images were randomly selected from each rice class, and their blur levels were compared using a boxplot to check for inconsistencies. This process helped identify whether any class contained significantly blurrier images that could introduce bias during training. The results showed that blur levels were generally similar across classes, suggesting that the dataset maintained sufficient visual quality for reliable model learning (Mahmud et al., 2021).

```python
# attribute: https://www.kaggle.com/code/merfarukyce/mnist-cnn-classification
# author: mer Faruk Yce
import cv2

def detect_blur(image_path, threshold=100):
    image = cv2.imread(image_path)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    laplacian_var = cv2.Laplacian(gray, cv2.CV_64F).var()
    return laplacian_var

# Sample a few images from each class to check blur
blur_data = []
for cls in classes:
    sample_paths = df[df['label'] == cls]['image'].sample(5)
    for path in sample_paths:
        blurriness = detect_blur(path)
        blur_data.append({'class': cls, 'blur': blurriness})

blur_df = pd.DataFrame(blur_data)
plt.figure(figsize=(10,6))
sns.boxplot(x='class', y='blur', data=blur_df)
plt.title('Blurriness by Class (Laplacian Variance)')
plt.show()
```

##**Analyze colour distribution**

The colour characteristics of the rice images were analyzed to support feature extraction for CNN model training (Zhang, Li and Wang, 2020). Sampled images from each rice variety were converted into numerical arrays using PIL, and the mean, median, and standard deviation of red, green, and blue channels were calculated. Histograms and a heatmap were used to visualize the color distributions across classes (Zhang, Li and Wang, 2020).

```python
# attribute: https://www.kaggle.com/code/thariqkhalid/simple-eda-image-formation-with-numpy
# author: Thariq Khalid

import matplotlib.pyplot as plt
import seaborn as sns

def analyze_color_distribution(df, sample_size=100):
    """Analyze color characteristics across rice varieties"""

    # Sample images from each class
    sampled_df = df.groupby('label').apply(lambda x: x.sample(min(sample_size, len(x)))).reset_index(drop=True)

    color_stats = []

    for _, row in sampled_df.iterrows():
        img = Image.open(row['image'])
        img_array = np.array(img)

        # Calculate colour statistics for each channel
        for channel, color in zip([0, 1, 2], ['Red', 'Green', 'Blue']):
            channel_data = img_array[:, :, channel].flatten()

            color_stats.append({
                'class': row['label'],
                'channel': color,
                'mean': np.mean(channel_data),
                'std': np.std(channel_data),
                'median': np.median(channel_data)
            })

    color_df = pd.DataFrame(color_stats)

    # Create 2 plots
    fig, axes = plt.subplots(1, 2, figsize=(15, 6))

    # Histogram of Red channel intensity
    red_data = color_df[color_df['channel'] == 'Red']
    for class_name in red_data['class'].unique():
        class_data = red_data[red_data['class'] == class_name]['mean']
        axes[0].hist(class_data, alpha=0.6, label=class_name, bins=20)
    axes[0].set_title('Red Channel Intensity Distribution by Class')
    axes[0].set_xlabel('Red Channel Intensity')
    axes[0].set_ylabel('Frequency')
    axes[0].legend()

    # Heatmap of average RGB per class
    pivot_data = color_df.pivot_table(index='class', columns='channel', values='mean')
    sns.heatmap(pivot_data, annot=True, cmap='YlOrRd', ax=axes[1])
    axes[1].set_title('Average Colour Intensity Heatmap')

    plt.tight_layout()
    plt.show()

    return color_df

# Run colour analysis
color_analysis = analyze_color_distribution(df)
```

##**Shape and Edge Analysis**

In this section, we analyze rice images in the frequency domain using the Fast Fourier Transform (FFT). The average frequency spectrum of each rice class is visualized to observe patterns in texture and structure. Metrics such as low-frequency energy, high-frequency energy, and the ratio between them are computed for each class. Low-frequency energy reflects the overall structure and smooth regions of the grain, while high-frequency energy captures fine details and edges. Comparing these metrics across classes helps identify textural and structural differences that may be important for CNN-based classification (Gonzalez, Woods and Eddins, 2021).

```python
# attribute: https://www.kaggle.com/code/arminzare1996/rice-classification-with-pytorch
# author: Armin Zare
from skimage import feature, filters

def simple_shape_analysis(df, sample_size=30):
    """Simplified shape and edge analysis for rice images"""

    metrics = []

    for class_name in df['label'].unique():
        samples = df[df['label'] == class_name].sample(min(sample_size, len(df[df['label'] == class_name])))

        for img_path in samples['image']:
            img = Image.open(img_path).convert('L')
            img_array = np.array(img)

            # Edge intensity
            edges = feature.canny(img_array, sigma=2)
            edge_intensity = np.sum(edges) / edges.size

            # Foreground area ratio
            thresh = filters.threshold_otsu(img_array)
            binary = img_array > thresh
            area_ratio = np.sum(binary) / binary.size

            # Circularity
            perimeter = np.sum(feature.canny(binary.astype(np.uint8)))
            circularity = (4 * np.pi * np.sum(binary)) / (perimeter ** 2) if perimeter > 0 else 0

            metrics.append({
                'class': class_name,
                'edge_intensity': edge_intensity,
                'area_ratio': area_ratio,
                'circularity': circularity
            })

    shape_df = pd.DataFrame(metrics)

    # Quick boxplot summary
    plt.figure(figsize=(15,4))
    for i, metric in enumerate(['edge_intensity','area_ratio','circularity']):
        plt.subplot(1,3,i+1)
        sns.boxplot(data=shape_df, x='class', y=metric)
        plt.title(metric.replace('_',' ').title())
        plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    return shape_df

# Run simplified shape analysis
shape_analysis = simple_shape_analysis(df)
```

##**Frequency Domain Analysis**

We visualize the average frequency spectrum per class and compute metrics such as low-frequency energy, high-frequency energy, and their ratio. This helps to understand textural patterns and structural differences across rice varieties.

```python
# attribute: https://gist.github.com/dbetchkal/e18b416c3fe3ae20bc081b0ac51531ba
# author: dbetchkal

def frequency_domain_analysis(df, sample_per_class=10):
    """Visualize average FFT spectrum per class (no low/high frequency plots)"""

    import matplotlib.pyplot as plt
    from PIL import Image
    import numpy as np

    class_names = df['label'].unique()
    num_classes = len(class_names)

    # Create subplot grid automatically
    n_cols = 3
    n_rows = int(np.ceil(num_classes / n_cols))
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(5*n_cols, 5*n_rows))
    axes = axes.flatten()  # Flatten in case of single row/column

    for i, class_name in enumerate(class_names):
        class_samples = df[df['label'] == class_name].sample(sample_per_class)
        avg_spectrum = None

        for _, row in class_samples.iterrows():
            img = Image.open(row['image']).convert('L')
            img_array = np.array(img)

            # Compute FFT and magnitude spectrum
            fft = np.fft.fft2(img_array)
            fft_shift = np.fft.fftshift(fft)
            magnitude_spectrum = np.log(np.abs(fft_shift) + 1)

            if avg_spectrum is None:
                avg_spectrum = magnitude_spectrum
            else:
                avg_spectrum += magnitude_spectrum

        avg_spectrum /= sample_per_class

        # Plot FFT spectrum per class
        im = axes[i].imshow(avg_spectrum, cmap='hot')
        axes[i].set_title(f'{class_name}\nFrequency Spectrum')
        axes[i].axis('off')
        plt.colorbar(im, ax=axes[i])

    # Remove any unused subplots
    for j in range(i+1, len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout()
    plt.show()

# Run frequency analysis
frequency_analysis = frequency_domain_analysis(df)
```

##**Heatmap of Class Similarity**

In this analysis, we made a heatmap to show how visually similar the rice classes are using colour histogram features. Each rice class was represented by the average colour-histogram vector of its images, and we calculated the cosine similarity between these vectors. The heatmap shows that most of the rice classes are very similar to each other in colour, but Ipsala stands outit's only strongly similar to itself. This suggests that Ipsala is visually distinct from the other rice types, meaning a CNN model might find it easier to recognise, while the other classes may need more careful feature extraction to separate them (Zhang, Liu and Wang, 2022).

```python
# attribute: https://courses.cs.washington.edu/courses/cse473/19wi/notes/Vision0-19.pdf
# author: University of Washington (Computer Vision course)

from sklearn.metrics.pairwise import cosine_similarity
import cv2

def plot_similarity_heatmap(df, sample_size=50):
    """Plot Heatmap of Inter-Class Visual Similarity"""

    features = []
    labels_list = []

    for class_name in df['label'].unique():
        class_samples = df[df['label'] == class_name].sample(min(sample_size, len(df[df['label'] == class_name])))

        for _, row in class_samples.iterrows():
            img = Image.open(row['image'])
            img_array = np.array(img)

            hist_features = []
            for channel in range(3):
                hist = cv2.calcHist([img_array], [channel], None, [32], [0, 256])
                hist_features.extend(hist.flatten())

            features.append(hist_features)
            labels_list.append(class_name)

    features = np.array(features)
    unique_classes = df['label'].unique()

    # Compute class mean features
    class_means = []
    for class_name in unique_classes:
        class_mask = np.array(labels_list) == class_name
        class_mean = features[class_mask].mean(axis=0)
        class_means.append(class_mean)

    similarity_matrix = cosine_similarity(class_means)

    # Plot heatmap
    plt.figure(figsize=(10, 8))
    sns.heatmap(similarity_matrix, annot=True, fmt='.3f', cmap='YlOrRd',
                xticklabels=unique_classes, yticklabels=unique_classes)
    plt.title('Inter-Class Visual Similarity Matrix')
    plt.tight_layout()
    plt.show()

    return similarity_matrix

# Run heatmap
similarity_matrix = plot_similarity_heatmap(df)
```

#**Confusion Matrix for Rice Classification and report**

After training the CNN, a confusion matrix was generated to analyse the model'S performance for each rice class. The matrix displayed how often each rice variety was correctly classified and where misclassifications occurred. This allowed deeper interpretation beyond accuracy scores and helped identify classes with similar visual characteristics that may be harder for the model to distinguish

```python
# attribute: https://www.kaggle.com/code/arthurtok/image-classification-evaluation
# author: Arthur Tok

from sklearn.metrics import confusion_matrix, classification_report

def evaluate_model(y_true, y_pred, class_names):
    """
    Plot confusion matrix and print classification report.

    Parameters:
    y_true : list or array - True labels
    y_pred : list or array - Predicted labels
    class_names : list - Names of the classes
    """

    # Ensure 1D numpy arrays of strings
    y_true = np.array(y_true).astype(str)
    y_pred = np.array(y_pred).astype(str)

    # Compute confusion matrix
    cm = confusion_matrix(y_true, y_pred, labels=class_names)

    # Plot confusion matrix heatmap
    plt.figure(figsize=(8,6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names)
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.title('Confusion Matrix')
    plt.show()

    # Print classification report
    report = classification_report(y_true, y_pred, target_names=class_names)
    print("Classification Report:\n")
    print(report)

    return cm, report

# Example usage:
y_true = ['Arborio', 'Basmati', 'Jasmine', 'Arborio', 'Basmati']
y_pred = ['Arborio', 'Basmati', 'Basmati', 'Arborio', 'Jasmine']
class_names = ['Arborio', 'Basmati', 'Jasmine']

cm, report = evaluate_model(y_true, y_pred, class_names)
```

```python
# attribute: https://www.kaggle.com/code/arminzare1996/rice-classification-with-pytorch
# author: Armin Zare

import numpy as np
import matplotlib.pyplot as plt

def plot_misclassified_images(model, x_val, y_true, class_names, num_images=5):
    """
    Plots misclassified images from validation set.

    Parameters:
    - model: trained Keras model
    - x_val: NumPy array of validation images
    - y_true: true labels (list, NumPy array, or one-hot encoded)
    - class_names: list of class names
    - num_images: number of misclassified images to display
    """

    # Convert y_true to NumPy array if it is a list
    y_true = np.array(y_true)

    # If y_true is one-hot encoded, convert to class indices
    if len(y_true.shape) > 1 and y_true.shape[1] > 1:
        y_true = np.argmax(y_true, axis=1)

    # Get predictions and convert to class indices
    y_pred_prob = model.predict(x_val)
    y_pred = np.argmax(y_pred_prob, axis=1)

    # Find misclassified indices
    misclassified_idx = np.where(y_true != y_pred)[0]

    if len(misclassified_idx) == 0:
        print("No misclassified images found!")
        return

    # Limit to num_images
    misclassified_idx = misclassified_idx[:num_images]

    # Plot misclassified images
    plt.figure(figsize=(15, 5))
    for i, idx in enumerate(misclassified_idx):
        plt.subplot(1, num_images, i+1)
        plt.imshow(x_val[idx])
        plt.axis('off')
        plt.title(f"True: {class_names[y_true[idx]]}\nPred: {class_names[y_pred[idx]]}", fontsize=10)
    plt.suptitle("Misclassified Images", fontsize=16)
    plt.show()

# Example usage:
# x_val, y_val_true already prepared from validation_generator
class_names = ['Basmati', 'Ipsala', 'Jasmine', 'Karacadag', 'Rozana']  # your rice classes
y_val_true = np.array(y_val_true)  # ensure it's a NumPy array
plot_misclassified_images(model, x_val, y_val_true, class_names, num_images=5)
```

```python
# attribute: https://www.analyticsvidhya.com/blog/2022/04/master-image-augmentation-with-widely-used-python-libraries/
# author: Devashree

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

# Set image size (match your generator)
IMG_HEIGHT, IMG_WIDTH = 50, 50
NUM_CLASSES = len(df['label'].unique())

# Build improved CNN model
improved_model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(IMG_HEIGHT, IMG_WIDTH, 3)),
    MaxPooling2D((2,2)),

    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D((2,2)),

    Conv2D(128, (3,3), activation='relu'),
    MaxPooling2D((2,2)),

    Flatten(),
    Dense(256, activation='relu'),
    Dropout(0.5),

    Dense(NUM_CLASSES, activation='softmax')
])

# Compile the model
improved_model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

improved_model.summary()

# Set callbacks
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True,
    verbose=1
)

checkpoint = ModelCheckpoint(
    'best_improved_rice_model.h5',
    monitor='val_accuracy',
    save_best_only=True,
    verbose=1
)

# Train the improved model
history = improved_model.fit(
    train_generator,
    validation_data=validation_generator,
    epochs=30,
    class_weight=class_weights,   # if you computed them earlier
    callbacks=[early_stopping, checkpoint],
    verbose=1
)
```
